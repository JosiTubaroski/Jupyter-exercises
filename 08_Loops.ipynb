**While**

```python
#O Loop while requer dois cuidados
# 1 - O Loop somente é executado se a condição for true
# 2 - Dentro do Loop a condição obrigatoriamente tem que passar o false ou break ser executado, para que o Loop pare.

contador = 0

while contador <= 15:
    
    print(contador)
    contador = contador + 1
```

```python
#A exemplo do IF / ELSE, quando o ELSE é executado após o return False do If, podemos também utilizar else com while
#o Loop executado enquanto a condição do while for verdadeira e depois entra no else

x = 0

while x < 10:
    print(x)
    x +=1
    
else:
    print('Loop concluido')
```

**Pass e Break**

```python
#São conjuntos que atuam juntos com o loop e while e dão maior flexibilidade, pois podem interromper o fluxo antes da condição 
#principal tarnar-se falsa
```

```python
#Observe a condição verdadeira
i = 0
while True:
    print('A condição continua verdadeira. I vale: ' + str(i))
    i = i + 1
    
    if i < 20:
        pass
    else:
        break
```

```python
i = 0
while i < 1000:
    if i == 20:
        break
    else:
        pass
    print(i)
    i = i + 1
```

```python
#Podemos utilizar o PASS para imprimir números ímpares, por exemplo

indice = 2
while indice < 30:
    if indice % 2 == 0:
        pass
    else:
        print(indice)
        
    indice = indice + 1
```
