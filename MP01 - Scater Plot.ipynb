### Aumentando a DISTANCIA entre os pontos (pares)

```python
x = []
y = []

for i in range(-50,100,5):
    x.append(i)
    y.append(i)
   
```

```python
print(x,y)
```

```python
plt.scatter(x,y)
```

### Modificando o tamanho dos pares

```python
plt.scatter(x,y, s=100)
```

```python
plt.scatter(x,y, s=5)
```

### Alterando a cor dos pares

```python
plt.scatter(x,y, s=30, c='green')
```

```python
plt.scatter(x,y, s=30, c='yellow')
```

### Alterando o marcador dos pares

```python
plt.scatter(x,y, s=100, c='purple', marker='^')
```

```python
plt.scatter(x,y, s=100, c='gray', marker='o')
```

### Gráfico multi linhas

```python
x = []
y = []
y2 = []
y3 = []

for i in range(-50,100,5):
    x.append(i)
    y.append(i)
    y2.append(i * 2)
    y3.append(i + 50)
```

```python
plt.scatter(x,y, s=100, c='gray', marker='o')
plt.scatter(x,y2, s=30, c='purple', marker='^')
plt.scatter(x,y3, s=100, c='blue', marker='+')
```
