# Filtros Multiplos

```python
import numpy as np
import pandas as pd
```

```python
colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration', 'actors_list']

file="C:\\projetos\\movies.csv"
filmes=pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns
```

```python
#Filtrando com duração superior a 180 minutos

filmes[filmes.duration >= 180]
```

Entendendo operadores logicos:

```python
#Operador AND

print(True and True)
print(True and False)
print(False and True)
print(False and False)
```

```python
#Operador OR

print(True or True)
print(True or False)
print(False or True)
print(False or False)
```

Regras de utilização de múltiplos filtros no pandas:
    
    * use & em vez de and
    * use | em vez de or
    * adicione parenteses em volta de cada critério especifico, por ordem

### Trazer somente os filtros do genero drama E com duração superior a 200 minutos

```python
dramas_longos = filmes[(filmes.duration >= 200) & (filmes.genre == 'Drama')]
```

```python
dramas_longos
```

### Trazer somente os filmes do genre drama OU com duração superior a 200 minutos

```python
#Trocando a condição AND por OR

dramas_longos = filmes[(filmes.duration >= 200) | (filmes.genre == 'Drama')]
dramas_longos
```

### Trazer os filmes dos gêneros Crime, Drama ou Action

```python
filmes[(filmes.genre == 'Crime') | (filmes.genre == 'Drama') | (filmes.genre == 'Action')].head(10)
```

### Utilizando o método ISIN

```python
filmes[filmes.star_rating.isin([9.0,9.1,9.2,9.3])]
```

```python
filmes[filmes.genre.isin(['Crime', 'Drama', 'Action'])].head(10)
```
