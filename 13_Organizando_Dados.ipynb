# O obejetivo é retirar a virgula e inverter a posição dos nomes

```python
import re

match = re.compile(r'"(.*)(,)(\s.*)"')
match.sub(r'"\3 \1"', linhas[1])
```

```python
#Criando uma funcao que remove a virgula nomes e troca a posicao
```

```python
def tratar_nome(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    return match.sub(r'"\3 \1"', linha)
```

```python
#Vamos testar a funcao na Linha 1

tratar_nome(linhas[1])
```

```python
linha1_tratada = tratar_nome(linhas[1])
```

```python
linha1_separada = linha1_tratada.split(",")
```

```python
print(colunas)
print(linha1_tratada)
print(linha1_separada)
```

```python
print(len(colunas), len(linha1_separada))
```

```python
#Aqui trataremos todas as linhas, excluindo a linha 0, que é o cabeçalho

for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    print(dado_tratado)
```

```python
#Aqui faremos a mesma coisa, porém vamos separar as linhas pela virgula e converter em uma lista

for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    dado_como_lista = dado_tratado.split(',')
    print(dado_como_lista)
    
```

```python
#Aqui faremos a mesma coisa dos passos anteriores, porem vamos criar uma lista para
#adicionar cada linha convertida em lista, fazendo uma lista aninhada

dados_tratados = []

for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    dado_como_lista = dado_tratado.split(',')
    dados_tratados.append(dado_como_lista)
```

# Utilizando o ENUMERATE. O enumerate retorna o Índice em que o elemento está e tambem o seu valor!

```python
frutas = ['mamão', 'melão', 'melancia']

for indice, fruta in enumerate(frutas):
    print('Fruta:', fruta, 'no indice', indice)
```

```python
# 1 - Pego cada linha da variavel dados_tratados e coloco em passageiro
# 2 - Faço um for em passageiro e pego o indice e o restante dos dados
# 3 - Vou na variavel coluna e pego a coluna no respectivo indice

for passageiro in dados_tratados:
    for indice, dado in enumerate(passageiro): #Enumerate retorna o INDICE e VALOR(atribuição MULTIPLA)
        coluna = colunas[indice]
        print(coluna,dado)
```

```python
lista_idades = []

for idade in titanic_dados["Age"]:
    if idade != -1:
        lista_idades.append(idade)
```

```python
def media(lista):
    soma_total = sum(lista)
    quantidade = len(lista)
    return soma_total / quantidade
    
```

```python
print(media(lista_idades))
```

```python
media_idades = media(lista_idades)
```
