### Renomeando colunas com Pandas

```python
filmes.columns
```

```python
filmes.rename(columns = {'star rating':'star_rating', 'content rating':'content_rating'})
```

A transformação não vai ocorrer, pois estamos modificando o dataframe original e o Pandas não permite essa operação, a não ser que ela seja explicita, ou seja, devemos forçar a alteração

  1 - Atribuir o resultado a uma nova variavel, ou seja um novo DataFrame
  
  2 - Utilizar o arqumento Inplace=True forçando a alteração

```python
filmes.rename(columns = {'star rating':'star_rating', 'content rating':'content_rating'},inplace=True)
```

```python
filmes.columns
```

Podemos também passar a lista de colunas EXPLICITAMENTE

# Real World

É muito comum termos arquivos com várias colunas e operações de renomear uma a uma fica inviável. Podemos então, utilizar 
funções do Python para trabalharmos com as colunas.

```python
file="C:\\projetos\\movies.csv"
filmes = pd.read_csv(file)
filmes.head()

filmes.columns = filmes.columns.str.replace(' ', '_')
filmes.columns
```

# Removendo Colunas do Dataframe

```python
#Por boa prática, ao realizar alguma alteração no Dataframe, devemos atribuir a outra variável,
#assim é possivel rastrear as mudanças e também voltar ao último passo.
#Aqui também podemos utilizar o inplace=true

#Axis = 1 - Colunas / Axis = 0 - Linhas

#filmes01 - foi criado sem a coluna content_

filmes01 = filmes.drop('content_rating', axis=1)
filmes01.head()
```

```python
#Removendo mais de uma coluna

filmes02 = filmes01.drop(['duration', 'actors_list'], axis=1)
filmes02.head()
```

```python
filmes03 = filmes02.drop([0, 1], axis=0)
filmes03.head()
```
