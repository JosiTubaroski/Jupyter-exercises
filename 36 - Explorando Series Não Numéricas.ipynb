### Aprofundando em séries

```python
import pandas as pd
import numpy as np
```

```python
colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration', 'actors_list']

file="C:\\projetos\\movies.csv"
filmes = pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns
```

```python
filmes.head()
```

```python
#Examinando os tipos
filmes.dtypes
```

### Explorando séries não numéricas

```python
# Contando valores não nulos, unique e a frequencia do valor mais comum
filmes.genre.describe()
```

```python
# Vamos verificar a frequencia dos valores

filmes.genre.value_counts()
```

```python
#Podemos exibir a porcentagem

filmes.genre.value_counts(normalize=True)
```

### Como podemos ver, value_counts() retorna uma série, e tal pode ser aproveitada com todos os métodos

```python
type(filmes.genre.value_counts())
```

```python
#Um exemplo simples é o head()

filmes.genre.value_counts().head(10)
```

```python
#Ou o tail()

filmes.genre.value_counts().tail(10)
```

### unique() e nunique()

```python
#O método unique mostra os valores da série sem repetir

filmes.genre.unique()
```

```python
#O método nunique() mostra a quantidade de valores unicos
filmes.genre.nunique()
```
