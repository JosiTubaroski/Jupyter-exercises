```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = []
y = []
y2 = []
```

```python
for num in range(-200,201,10):
    x.append(num)
    y.append(num)
    y2.append(2 * num)
```

```python
#Perceba a diferença entre o método: plot <> scatter

#plt.scatter(x,y, s=100, c='gray', marker='o')

plt.plot(x,y)
```

### Alterando a cor da linha

```python
plt.plot(x,y, c = "red")
```

### Alterando os marcadores - style

```python
plt.plot(x,y,c = "blue", ls = "--")
```

```python
plt.plot(x,y,c = "blue", ls = ":")
```

```python
plt.plot(x,y,c = "blue", ls = "-.")
```

### Alterando o tamanho (espessura)

```python
plt.plot(x,y,c = "blue", ls = "-.", lw = 10)
```

### Inserindo marcadores

```python
plt.plot(x,y,c = "blue", ls = "-.", lw = 1, marker = "o")
```

### Alterando o tamanho dos marcadores

```python
plt.plot(x,y,c = "red", ls = "-", lw = 5, marker = "o", ms = 10)
```

```python
#Customizando markers
```

```python
plt.plot(x,y,c = "blue", ls = "-", lw = 5, marker = "o", ms = 20, fillstyle = "bottom")
```
