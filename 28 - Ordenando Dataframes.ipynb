# Ordenando Dataframes

```python
import numpy as np
import pandas as pd
```

```python
colunas_filmes = ['start_rating', 'title', 'content_rating', 'genre', 'duration', 'actors_list']

file="C:\\projetos\\movies.csv"
filmes = pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns
```

```python
filmes.head()
```

```python
#Ordenando por titulo
filmes.title.sort_values().head()
```

```python
#O retorno é uma série

A = filmes.title.sort_values()
type(A)
```

```python
A
```

```python
#Ordem descendente

filmes.title.sort_values(ascending=False).head(10)
```

```python
#Há outra forma de ordenar um DataFrame. Desta forma, o retorno é um DataFrame

filmes.sort_values('title').head()
```

```python
#Podemos também ordenar de forma descendente

filmes.sort_values('title', ascending=False).head()
```

```python
#Apenas ordenar não altera o DataFrame original

filmes.head()
```

```python
#Ordenando por classificação e duração

filmes.sort_values(['genre', 'duration']).head(20)
```

# Filtrando um DataFrame

```python
#Vamos criar uma lista para saber os filmes com maior duração (Maior ou igual 200)

boleanos = []

for length in filmes.duration:
    if length >= 200:
        boleanos.append(True)
    else:
        boleanos.append(False)

```

```python
boleanos [:20]
```

```python
#Verificando o tamanho da lista

len(boleanos)
```

```python
type(boleanos)
```

Precisamos converter a lista em uma SERIE para aplicar como filtro de Dataframe Pandas

```python
#Convertendo a lista em serie

duracao = pd.Series(boleanos)
duracao.head()
```

```python
#Podemos usar a notação de colchetes para indexar ou para aplicar filtros de series boleanas

print(filmes['title'][:10])
```

```python
filmes[duracao]
```

```python
#Podemos ainda continuar a notação trazendo somente as colunas desejadas

filmes[duracao][['title', 'start_rating']]
```

Podemos simplificar o filtro conforme abaixo ou ainda escrevendo em List Comprehension

```python
#Forma simplificada

A = filmes.duration >= 180
A.head(20)
```

```python
filmes[A].head()
```

```python
#Escrevendo por List Comprehension

filmes[filmes.duration >= 180].head()
```

```python
#Podemos também retornar outra série com a coluna que desejamos

genero_maior_200 = filmes[filmes.duration >= 200].genre
```

```python
genero_maior_200
```

```python
#Podemos ainda utilizar a notação de colchetes para fazer a mesma coisa

genero_Maior_200 = filmes[filmes.duration >= 200]['genre']
```

```python
genero_Maior_200
```

```python
#Porém, por boa prática, utilizamos sempre o método LOC que nada mais é do que definir uma condição e filtrar por labels
#Veremos o ILOC em mais detalhes

filmes.loc[filmes.duration >= 200, 'genre']
```
